```python

import requests
from bs4 import BeautifulSoup
import csv
import urllib.parse
```

```python
WEBSITES = [
    'https://timesofindia.indiatimes.com',
    'https://www.livehindustan.com',
    'https://odishatv.in'
]
```

```python
def is_internal_link(url, base_url):
    return urllib.parse.urlparse(url).netloc == urllib.parse.urlparse(base_url).netloc
```

```python
def fetch_and_parse_url(url):
    response = requests.get(url)
    if response.status_code == 200:
        return BeautifulSoup(response.content, 'html.parser')
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None
```

```python
def extract_text(soup):
    paragraphs = soup.find_all('p')
    text_data = [p.get_text().strip() for p in paragraphs if p.get_text().strip()]
    return text_data
```

```python
def get_internal_links(soup, base_url):
    links = set()
    for a_tag in soup.find_all('a', href=True):
        href = a_tag['href']
        full_url = urllib.parse.urljoin(base_url, href)
        if is_internal_link(full_url, base_url):
            links.add(full_url)
    return links
```

```python
def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerow(['text'])  # Write header
        for row in data:
            csvwriter.writerow([row])
```

```python
def main():
    all_text_data = []

    for website in WEBSITES:
        print(f"Processing {website}")
        # Fetch and parse the main page
        main_soup = fetch_and_parse_url(website)
        if not main_soup:
            continue
        
        # Extract text from the main page
        text_data = extract_text(main_soup)
        
        # Get internal links from the main page
        internal_links = get_internal_links(main_soup, website)
        
        # Crawl and extract text from each internal link
        for link in internal_links:
            print(f"Fetching {link}")
            soup = fetch_and_parse_url(link)
            if soup:
                text_data.extend(extract_text(soup))
        
        # Add extracted text data to the overall list
        all_text_data.extend(text_data)
```

```python

if __name__ == "__main__":
    main()
```
